Исходные данные имеют следующую структуру:

перечень всех документов

Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
Примеры работы:

Введите команду:
p

Введите номер документа:
10006
Результат:
Владелец документа: Аристарх Павлов

Введите команду:
p

Введите номер документа:
12345
Результат:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def find_owner_by_doc_number(doc_number):
    for document in documents:
        if document['number'] == doc_number:
            return document['name']
    return None

def main():
    while True:
        command = input('Введите команду:...')
        if command == 'q':
            break
        elif command == 'p':
            doc_number = input('Введите номер документа:...')
            owner = find_owner_by_doc_number(doc_number)
            if owner:
                print(f'Владелец документа: {owner}')
            else:
                print('Документ не найден в базе')
        else:
            print('Неизвестная команда')

if __name__ == '__main__':
    main()
```

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def find_owner_by_doc_number(doc_number):
    for document in documents:
        if document['number'] == doc_number:
            return document['name']
    return None
def find_shelf_by_doc_number(doc_number):
    for shelf, documents in directories.items():
        if doc_number in documents:
            return shelf
    return None
######### Добавляем новую команду l
def list_documents():
    for doc in documents:
        shelf = find_shelf_by_doc_number(doc['number'])
        print(f'№: {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {shelf or "не найден"}')
##################
def main():
    while True:
        command = input('Введите команду:...')
        if command == 'q':
            break
        elif command == 'p':
            doc_number = input('Введите номер документа:...')
            owner = find_owner_by_doc_number(doc_number)
            if owner:
                print(f'Владелец документа: {owner}')
            else:
                print('Документ не найден в базе')
        elif command == 's':
            doc_number = input('Введите номер документа:...')
            shelf = find_shelf_by_doc_number(doc_number)
            if shelf:
                print(f'Документ хранится на полке: {shelf}')
            else:
                print('Документ не найден в базе') 
################### Добавляем вызов новой команды l        
        elif command == 'l':
            list_documents()
###################
        else:
            print('Неизвестная команда')

if __name__ == '__main__':
    main()
```

Пункт 4. Пользователь по команде "ads" может добавить новую полку
Примеры работы:

Введите команду:
ads

Введите номер полки:
10
Результат:
Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.

Введите команду:
ads

Введите номер полки:
1
Результат:
Такая полка уже существует. Текущий перечень полок: 1, 2, 3.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def find_owner_by_doc_number(doc_number):
    for document in documents:
        if document['number'] == doc_number:
            return document['name']
    return None
def find_shelf_by_doc_number(doc_number):
    for shelf, documents in directories.items():
        if doc_number in documents:
            return shelf
    return None
def list_documents():
    for doc in documents:
        shelf = find_shelf_by_doc_number(doc['number'])
        print(f'№: {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {shelf or "не найден"}')
######### Добавляем новую команду ads
def add_shelf(shelf_number):
    if shelf_number in directories:
        print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(directories.keys())}.')
    else:
        directories[shelf_number] = []
        print(f'Полка добавлена. Текущий перечень полок: {", ".join(directories.keys())}.')
##################
def main():
    while True:
        command = input('Введите команду:...')
        if command == 'q':
            break
        elif command == 'p':
            doc_number = input('Введите номер документа:...')
            owner = find_owner_by_doc_number(doc_number)
            if owner:
                print(f'Владелец документа: {owner}')
            else:
                print('Документ не найден в базе')
        elif command == 's':
            doc_number = input('Введите номер документа:...')
            shelf = find_shelf_by_doc_number(doc_number)
            if shelf:
                print(f'Документ хранится на полке: {shelf}')
            else:
                print('Документ не найден в базе')       
        elif command == 'l':
            list_documents()
################### Добавляем вызов новой команды ads
        elif command == 'ads':
            shelf_number = input('Введите номер полки:...')
            add_shelf(shelf_number)
###################
        else:
            print('Неизвестная команда')

if __name__ == '__main__':
    main()
```
